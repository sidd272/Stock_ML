```python
import pandas as pd
import numpy as np
```

```python
def sma(series,Period):
    return series.rolling(Period).mean()
```

```python
def wma(series,Period):
    weights = np.ones(Period) / Period
    return np.convolve(series, weights, mode='same')
```

```python
def ema(series, Period):
    alpha=2 / (Period + 1)
    w = np.empty(len(series))
    w[0] = 1
    for i in range(1, len(series)):
        w[i] = w[i - 1] * (1 - alpha) + alpha * series[i]
    return w
```

```python
def vwma(series,Period):
    tpv = np.cumprod(series * close, axis=0)
    cum_volume = np.cumsum(series, axis=0)
    vwma = tpv / cum_volume
    return vwma
```

```python
def rma(series,Period):
    rma = series.ewm(span=Period, min_periods=Period - 1).mean()
    return rma
```

```python
def change(open, high, low, close, period):
    return close - close[period - 1]
```

```python
def tr(open,high,low,close):
    return max(high-low,abs(high-close[1]),abs(low-close[1]))
```

```python
def ATR(high, low, close, period):
    tr = pd.DataFrame({'high': high, 'low': low, 'close': close})
    tr['tr1'] = tr['high'] - tr['low']
    tr['tr2'] = abs(tr['high'] - tr['close'].shift(1))
    tr['tr3'] = abs(tr['low'] - tr['close'].shift(1))
    
    tr['true_range'] = tr[['tr1', 'tr2', 'tr3']].max(axis=1)
    atr = tr['true_range'].rolling(window=period).mean()
    return atr
```

```python
def nz(x, y=None):
    if isinstance(x, np.generic):
        return x.fillna(y or 0)
    if x != x:
        if y is not None:
            return y
        return 0
    return x
```

```python
def Supertrend(data, periods=10, multiplier=3.0):
    src = (data["High"] + data["Low"]) / 2
    data['TR'] = ATR(data['High'], data['Low'], data['Close'], 14)
    atr2 = sma(data['TR'], periods)
    
    def nz(x, y):
        return np.where(pd.isna(x), y, x)
    
    up = src - (multiplier * atr2)
    up1 = nz(pd.Series(up).shift(1), up)
    up = np.where(data['Close'].shift(1) > up1, up, up1)
    
    dn = src + (multiplier * atr2)
    dn1 = nz(pd.Series(dn).shift(1), dn)
    dn = np.where(data['Close'].shift(1) < dn1, dn, dn1)
    
    trend = np.ones(len(data))  # Initialize as an array of ones
    for i in range(1, len(data)):
        trend[i] = trend[i - 1]
        if trend[i] == 1:
            if data['Close'][i] < up1[i]:
                trend[i] = -1
        else:
            if data['Close'][i] > dn1[i]:
                trend[i] = 1
    
    buySignal = (pd.Series(trend) == 1) & (pd.Series(trend).shift(1) == -1)
    sellSignal = (pd.Series(trend) == -1) & (pd.Series(trend).shift(1) == 1)
    
    return trend, buySignal, sellSignal,up,dn

```

```python
def MACD(close):
    fast_length = 12
    slow_length = 26
    src = close
    signal_length = 9
    sma_source = "EMA"
    sma_signal = "EMA"
    fast_ma = ema(src, fast_length)
    slow_ma =  ema(src, slow_length)
    macd = fast_ma - slow_ma
    signal = ema(macd, signal_length)
    hist = macd - signal
    return [macd,signal,hist]
```

```python
def rsi():
    rsiLengthInput = 14
    rsiSourceInput = close
    maTypeInput = "SMA"
    maLengthInput = 14
    bbMultInput = 2.0
    up = rma(max(change(rsiSourceInput), 0), rsiLengthInput)
    down = rma(-min(change(rsiSourceInput), 0), rsiLengthInput)
    rsi = 100 if down == 0 else  0 if up == 0 else 100 - (100 / (1 + up / down))
    return rsi
```

```python
def OpenCloseCross(data, delayOffset=0, basisType="SMA", basisLen=8):
    closeSeries = sma(data['Close'].shift(delayOffset), basisLen)
    openSeries = sma(data['Open'].shift(delayOffset), basisLen)
    xlong = data['Close'] > openSeries
    xshort = data['Close'] < openSeries
    return xlong, xshort
```

```python
def CenterOfLinearity(data, length=14):
    src = data['Close']
    COL=[]
    for i in range(0,len(src)-14):
        a = 0
        for j in range(1, length + 1):
            a = a + (j * (src.iloc[i+length - j] - src.iloc[i+j - 1]))
        COL.append(a)
    for i in range(0,14):
        COL.append(np.NaN)
    return COL
```

```python
def ADX_DI(data, period=14, threshold=20):
    data['TrueRange'] = data['High'] - data['Low']
    data['High_Close'] = abs(data['High'] - data['Close'].shift(1))
    data['Low_Close'] = abs(data['Low'] - data['Close'].shift(1))

    data['TrueRange'] = data[['TrueRange', 'High_Close', 'Low_Close']].max(axis=1)

    data['DirectionalMovementPlus'] = (data['High'] - data['High'].shift(1)).clip(lower=0).fillna(0)
    data['DirectionalMovementMinus'] = (data['Low'].shift(1) - data['Low']).clip(lower=0).fillna(0)

    data['SmoothedTrueRange'] = data['TrueRange'].rolling(window=period).mean()
    data['SmoothedDirectionalMovementPlus'] = data['DirectionalMovementPlus'].rolling(window=period).mean()
    data['SmoothedDirectionalMovementMinus'] = data['DirectionalMovementMinus'].rolling(window=period).mean()

    data['DIPlus'] = (data['SmoothedDirectionalMovementPlus'] / data['SmoothedTrueRange']) * 100
    data['DIMinus'] = (data['SmoothedDirectionalMovementMinus'] / data['SmoothedTrueRange']) * 100

    data['DX'] = (abs(data['DIPlus'] - data['DIMinus']) / (data['DIPlus'] + data['DIMinus'])) * 100
    data['ADX'] = data['DX'].rolling(window=period).mean()

    return data['ADX']
```
